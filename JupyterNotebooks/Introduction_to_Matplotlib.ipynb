# Introduction to `Matplotlib`
Working with data often requires being able to present that data in a way that illustrates trends or highlights important results.  As scientists, we are expected to provide plots and charts for nearly every piece of data we generate, especially when we're using that data to support a hypothesis.  As humans, we are also gifted with exceptional pattern-recognition abilities, which a savvy person can use to make their data compelling and easy to understand.

The `matplotlib` library is a powerful graphics tool in python which allows us to present our data in crisp, clean formats and provides a ton of customization options.  It contains many submodules, each with their own specific uses and functions.  Rest assured that these submodules are able to access the full array of `matplotlib` functionality, even if your own code is only pulling a small portion at any given time.

At this stage in the workshop, I'll assume you are already somewhat familiar with a lot of the shortcuts and syntax patterns in use, so many things won't be explicitly explained.

````python
import matplotlib.pyplot as plt
````

You may recall earlier discussions of "Object-Oriented Programming", where we are treating our programs like a series of objects that hold data and have their own functions within.  The `matplotlib` library is a fantastic example of this principle at work, with the core objects being the `figure()` and `axes()`.

First, we'll need to initialize a figure object and a set of axes in that figure.  There are a lot of keywords that can be included in these initializations, but for now we'll start with the size and resolution.

Resolution is in dots per inch.  75dpi is considered web-quality, while 300dpi is print-quality.  It's usually recommended that you create figures at higher resolutions than you'll need at the end, as it's easier to maintain quality of an image when shrinking it than when expanding it.  Starting with higher resolution allows for some adjustment room later on.  It costs you nothing but drive storage space and perhaps some adjustments to your font sizes along the way.

```python
fig = plt.figure(figsize = (<width>,<height>), # in inches
                 dpi = <resolution>) # in dpi
```

Now that we have a basic figure object, we need to add a set of axes to actually plot our data on.  In this first example, we'll use the `.add_subplot()` function, shown below.

```python
ax = fig.add_subplot(<n_rows>,<n_cols>,<cell_number>)
```

The `cell_number` value above refers to which position in the grid defined by `n_rows` and `n_cols`, such that the top-left position is 1, increasing across the row first, then the column.

So, a $3x2$ grid would look be numbered like this:

```
-------------------
|  1  |  2  |  3  |
-------------------
|  4  |  5  |  6  |
-------------------
```

The cell below has this layout shown, except that it doesn't bother making axes at position 4, which therefore remains blank.

````python
fig = plt.figure(figsize=(3,2), # 3in by 2in figure.
                 dpi=300) # 300 dpi

ax = fig.add_subplot(2,3,1)
ax = fig.add_subplot(2,3,2)
ax = fig.add_subplot(2,3,3)
# ax = fig.add_subplot(2,3,4)
ax = fig.add_subplot(2,3,5)
ax = fig.add_subplot(2,3,6)
````

Now that we've done this, we can see the x and y ticks are overlapping other axes, making the figure look really bad.  To solve this, we'll use the `.tight_layout()` function on the figure object once everything else is done.

It's important to do this just before displaying or saving your figure, but after all the other plotting and formatting of axes is done.

````python
fig = plt.figure(figsize=(3,2), # 3in by 2in figure.
                 dpi=300) # 300 dpi

ax = fig.add_subplot(2,3,1)
ax = fig.add_subplot(2,3,2)
ax = fig.add_subplot(2,3,3)
# ax = fig.add_subplot(2,3,4)
ax = fig.add_subplot(2,3,5)
ax = fig.add_subplot(2,3,6)

fig.tight_layout()
````
