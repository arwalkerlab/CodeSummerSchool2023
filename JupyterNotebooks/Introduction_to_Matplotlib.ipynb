# Introduction to `Matplotlib`
Working with data often requires being able to present that data in a way that illustrates trends or highlights important results.  As scientists, we are expected to provide plots and charts for nearly every piece of data we generate, especially when we're using that data to support a hypothesis.  As humans, we are also gifted with exceptional pattern-recognition abilities, which a savvy person can use to make their data compelling and easy to understand.

The `matplotlib` library is a powerful graphics tool in python which allows us to present our data in crisp, clean formats and provides a ton of customization options.  It contains many submodules, each with their own specific uses and functions.  Rest assured that these submodules are able to access the full array of `matplotlib` functionality, even if your own code is only pulling a small portion at any given time.

At this stage in the workshop, I'll assume you are already somewhat familiar with a lot of the shortcuts and syntax patterns in use, so many things won't be explicitly explained.

You may recall earlier discussions of "Object-Oriented Programming", where we are treating our programs like a series of objects that hold data and have their own functions within.  The `matplotlib` library is a fantastic example of this principle at work, with the core object being the `figure()`.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(3,2),
                 dpi=300)
ax = fig.add_subplot(1,1,1)
ax.plot([1,3,4,2,9,6,2,3,4,5,2,1,4,3,2,5,2])
plt.show()
```
